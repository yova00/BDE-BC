```python
#Procedimiento para analizar la calidad de una serie en especifico
import pandas as pd

# Ruta del archivo txt
archivo = 'IPC.txt'
df_ipc = pd.read_csv(archivo, delimiter='\t')

df_ipc.head(5)

```

```python
import pandas as pd

df_ipc['IPC'] = df_ipc['IPC'].str.replace(',', '.') #Reemplazar comas por puntos y eliminar caracteres no numéricos
df_ipc['Petroleo_1'] = df_ipc['Petroleo_1'].str.replace(',', '.')
df_ipc['Cobre_6'] = df_ipc['Cobre_6'].str.replace(',', '.')
df_ipc['Desempleo_12'] = df_ipc['Desempleo_12'].str.replace(',', '.')

df_ipc['IPC'] = df_ipc['IPC'].str.replace('[^0-9.-]', '') #Eliminar caracteres no numéricos por espacios en blanco
df_ipc['Petroleo_1'] = df_ipc['Petroleo_1'].str.replace('[^0-9.-]', '')
df_ipc['Cobre_6'] = df_ipc['Cobre_6'].str.replace('[^0-9.-]', '')
df_ipc['Desempleo_12'] = df_ipc['Desempleo_12'].str.replace('[^0-9.-]', '')

df_ipc['IPC'] = pd.to_numeric(df_ipc['IPC'], errors='coerce') #Convertir las columnas a tipo numérico
df_ipc['Petroleo_1'] = pd.to_numeric(df_ipc['Petroleo_1'], errors='coerce')
df_ipc['Cobre_6'] = pd.to_numeric(df_ipc['Cobre_6'], errors='coerce')
df_ipc['Desempleo_12'] = pd.to_numeric(df_ipc['Desempleo_12'], errors='coerce')

df_ipc.head(5)
```

```python
#verificar la calidad de los datos

#filtro por fecha de inicio y fin
df_ipc = df_ipc.loc[(df_ipc['Períodos'] >= '2011-03-01') & (df_ipc['Períodos'] <= '2022-12-01')]


if df_ipc.isnull().any().any(): #Verificar integridad de los datos
    print("El DataFrame contiene valores faltantes.\n")
missing_values = df_ipc.isnull().sum()
if missing_values.any():
    print(f"{missing_values.sum()} valores faltantes en total.\n")

#verificar valores atípicos en la columna IPC
mean_ipc = df_ipc['IPC'].mean() #Calcular la media y la desviación estándar
std_ipc = df_ipc['IPC'].std()
outliers_ipc = df_ipc.loc[(df_ipc['IPC'] > mean_ipc + 3 * std_ipc) | (df_ipc['IPC'] < mean_ipc - 3 * std_ipc)]
if not outliers_ipc.empty:
    print("El DataFrame contiene valores atípicos en la columna IPC:\n", outliers_ipc, "\n")

#valores negativos en la columna IPC
negativos_ipc = df_ipc.loc[df_ipc['IPC'] < 0]
if not negativos_ipc.empty:
    print("El DataFrame contiene valores negativos en la columna IPC:\n", negativos_ipc, "\n")

#verificar valores atípicos en la columna Petroleo_1
mean_petroleo = df_ipc['Petroleo_1'].mean() #Calcular la media y la desviación estándar
std_petroleo = df_ipc['Petroleo_1'].std()
outliers_petroleo = df_ipc.loc[(df_ipc['Petroleo_1'] > mean_petroleo + 3 * std_petroleo) | (df_ipc['Petroleo_1'] < mean_petroleo - 3 * std_petroleo)]
if not outliers_petroleo.empty:
    print("El DataFrame contiene valores atípicos en la columna Petroleo_1:\n", outliers_petroleo, "\n")


#verificar valores atípicos en la columna Cobre_6
mean_cobre = df_ipc['Cobre_6'].mean() #Calcular la media y la desviación estándar
std_cobre = df_ipc['Cobre_6'].std()
outliers_cobre = df_ipc.loc[(df_ipc['Cobre_6'] > mean_cobre + 3 * std_cobre) | (df_ipc['Cobre_6'] < mean_cobre - 3 * std_cobre)]
if not outliers_cobre.empty:
    print("El DataFrame contiene valores atípicos en la columna Cobre_6:\n", outliers_cobre, "\n")

#verificar valores atípicos en la columna Desempleo_12
mean_desempleo = df_ipc['Desempleo_12'].mean() #Calcular la media y la desviación estándar
std_desempleo = df_ipc['Desempleo_12'].std()
outliers_desempleo = df_ipc.loc[(df_ipc['Desempleo_12'] > mean_desempleo + 3 * std_desempleo) | (df_ipc['Desempleo_12'] < mean_desempleo - 3 * std_desempleo)] # mayor o menor a 3 desviaciones estándar
if not outliers_desempleo.empty:
    print("El DataFrame contiene valores atípicos en la columna Desempleo:\n", outliers_desempleo, "\n")
```

```python
#informe de calidad de la serie
print("Informe de calidad de la serie:\n")
print("Cantidad de filas y columnas:", df_ipc.shape,"\n")
print("Tipos de datos:\n", df_ipc.dtypes,"\n")
print("Valores faltantes:\n", df_ipc.isnull().sum(),"\n")
print("Valores atípicos:\n", df_ipc.describe(),"\n")
```
